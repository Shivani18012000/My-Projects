```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

books= pd.read_csv("data/BX-Books.csv",encoding='Latin-1',sep=";",error_bad_lines=False)
```

```python
books=books[["ISBN","Book-Title",	"Book-Author"	,"Year-Of-Publication",	"Publisher","Image-URL-L"]]
books.head(2)
```

```python
books=books.rename(columns={
    "Book-Title" :"Title",	"Book-Author":"Author"	,"Year-Of-Publication":"YOP","Image-URL-L":"url"
})
books.head(2)
```

```python
users=pd.read_csv("data/BX-Users.csv",encoding='Latin-1',sep=";",error_bad_lines=False)
users.head()
```

```python
ratings=pd.read_csv("data/BX-Book-Ratings.csv",encoding='Latin-1',sep=";",error_bad_lines=False)
ratings.head()
```

```python
x=ratings['User-ID'].value_counts()>200
```

```python
y=x[x].index
ratings=ratings[ratings['User-ID'].isin(y)]
```

```python
ratings_books=ratings.merge(books,on="ISBN")
ratings_books.head()
```

```python
ratings_books.shape
```

```python
num_ratings=ratings_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
num_ratings=num_ratings[num_ratings['Book-Rating']>=50]
```

```python
ratings_books_final=ratings_books.merge(num_ratings,on='Title')
ratings_books_final.rename(columns={
    "Book-Rating_y":"NumRatings"
},inplace=True)
```

```python
ratings_books_final.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
ratings_books_final.rename(columns={
    "Book-Rating_x":"rating"
},inplace=True)
```

```python
ratings_books_final
```

```python
pivot_books=ratings_books_final.pivot_table(columns='User-ID',index='Title',values='rating')
```

```python
pivot_books
```

```python
pivot_books.fillna(0,inplace=True)
pivot_books
```

```python
from scipy.sparse import csr_matrix
book_sparse= csr_matrix(pivot_books)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model= NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(pivot_books.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range (len(suggestion)):
    print(pivot_books.index[suggestion[i]])
```

```python
books_names=pivot_books.index
```

```python
import pickle
pickle.dump(model,open('modules/model.pkl','wb'))
pickle.dump(books_names,open('modules/books_names.pkl','wb'))
pickle.dump(ratings_books_final,open('modules/ratings_books_final.pkl','wb'))
pickle.dump(pivot_books,open('modules/pivot_books.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id=np.where(pivot_books.index==book_name)[0][0]
    distance,suggestion=model.kneighbors(pivot_books.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range (len(suggestion)):
        books=pivot_books.index[suggestion[i]]
        for j in books :
            print(j)

recommend_book('Harry Potter and the Goblet of Fire (Book 4)')
```
